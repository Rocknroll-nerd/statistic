```
df = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
df
```

```
df.Date = pd.to_datetime(df.date)
df.set_index('date', inplace = False)
```

```
df_new= df[df['location']=='Italy']['new_deaths']
```

```
dt=df[df['location']=='Russia']
#df_new = df[df['location']=='Aruba']
```

```
new = df[df['location']=='Italy'].date
```

```
plt.rcParams.update({'font.size': 32})
```

```
plt.figure(figsize=(30,10))
plt.plot(new, df_new, lw=2)
plt.xticks(rotation=70)
plt.xlabel('YYYY-MM-DD')
plt.axes().xaxis.set_major_locator(ticker.MultipleLocator(10))
plt.ylabel('Number of deaths in Italy')
plt.title('Italy mortality due to COVID-19')
```
